#6. Ranking por medalhas totais.

```python
# somando todas as medalhas de ouro da olimpiada de 2021
medalhas_ouro = df_medalhas['Gold'].sum()
```

```python
# somando todas as medalhas de prata da olimpiada de 2021
medalhas_prata = df_medalhas['Silver'].sum()
```

```python
# somando todas as medalhas de bronze da olimpiada de 2021
medalhas_bronze = df_medalhas['Bronze'].sum()
```

```python
# ordenando a tabela pela coluna Total do maior para o menor 
df_medalhas = df_medalhas.sort_values(by=['Total'], ascending=False)
```

```python
# criando uma nova linha com os totais
df_total_medalhas= df_medalhas.append({'Rank': 'Geral', 'Team/NOC': 'Total', 'Gold': medalhas_ouro, 'Silver': medalhas_prata, 'Bronze': medalhas_bronze, 'Total': medalhas_ouro+medalhas_prata+medalhas_bronze}, ignore_index = True)
```

```python
# localizando a linha de totais para poder mostrar posteriormente ao usuário
df_total_medalhas = df_total_medalhas.loc[(df_total_medalhas['Total'] == 1080)]
```

```python
# visualização do usuário
df_total_medalhas
```

```python
# Comparação de Medalhas totais da Olimpíadas
df_total_medalhas.groupby("Total")["Gold", "Silver", "Bronze"].sum().plot.bar(title= "Comparação de Medalhas totais da Olimpíadas")
plt.xlabel("Total")
plt.ylabel("Quantidades")
plt.xticks(rotation='horizontal'); #escritas do eixo x na horizontal
```
